# One-dimensional clustering

Clustering in many dimensions is a difficult problem with many proposed solutions. In Python, many solutions have been implemented in [scikit-learn](http://scikit-learn.org/stable/modules/clustering.html). In one-dimension, however, the problem is more tractable, resulting in solutions specific to the one-dimensional case. How do the more sophisticated and general clustering algorithms do in one-dimension compared to the more specific solutions?

```python
# Set up some general imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Test data

To test, we need a vector. We'll use `scikit-learn`'s dataset generating abilities to make a vector with samples from 3 Gaussians.

```python
from sklearn.datasets import make_blobs
n_clusters = 3
data, labels = make_blobs(n_samples=40, n_features=1, centers=n_clusters)
data.ravel()  # Make one-dimensional

plt.figure(figsize=(10,2))
plt.subplot(121)
plt.scatter(data, np.zeros_like(data))
plt.yticks(())
plt.subplot(122)
plt.scatter(data, labels, c=labels)
plt.yticks(());
```

Since we know what Gaussian produced each point, we can separate them in the right subplot, and color them approprately; in actuality, though, what the algorithm will receive is the unlabeled data in the left subplot.

You might want to run the above cell a few times to get data that is reasonably separate, but still with some overlap.

# Jenks Natural Breaks

The most common algorithm for clustering in one dimension is called Jenks Natural Breaks. I am not well versed in the ideas behind the algorithm; interested readers are directed to [Wikipedia](https://en.wikipedia.org/wiki/Jenks_natural_breaks_optimization). At the moment, there is no well-packaged and NumPy-aware implementation of Jenks in Python. The best available implementation is in [this Gist](https://gist.github.com/llimllib/4974446), which is explained in a [blog post](http://macwright.org/2013/02/18/literate-jenks.html) and copied below.

```python
def jenks_matrices_init(data, n_classes):
    #fill the matrices with data+1 arrays of n_classes 0s
    lower_class_limits = []
    variance_combinations = []
    for i in xrange(0, len(data)+1):
        temp1 = []
        temp2 = []
        for j in xrange(0, n_classes+1):
            temp1.append(0.)
            temp2.append(0.)
        lower_class_limits.append(temp1)
        variance_combinations.append(temp2)

    inf = float('inf')
    for i in xrange(1, n_classes+1):
        lower_class_limits[1][i] = 1.
        variance_combinations[1][i] = 0.
        for j in xrange(2, len(data)+1):
            variance_combinations[j][i] = inf

    return lower_class_limits, variance_combinations
 
def jenks_matrices(data, n_classes):
    lower_class_limits, variance_combinations = jenks_matrices_init(data, n_classes)
 
    variance = 0.0
    for l in xrange(2, len(data)+1):
        sum = 0.0
        sum_squares = 0.0
        w = 0.0
        for m in xrange(1, l+1):
            # `III` originally
            lower_class_limit = l - m + 1
            val = data[lower_class_limit-1]
 
            # here we're estimating variance for each potential classing
            # of the data, for each potential number of classes. `w`
            # is the number of data points considered so far.
            w += 1
 
            # increase the current sum and sum-of-squares
            sum += val
            sum_squares += val * val
 
            # the variance at this point in the sequence is the difference
            # between the sum of squares and the total x 2, over the number
            # of samples.
            variance = sum_squares - (sum * sum) / w
 
            i4 = lower_class_limit - 1
 
            if i4 != 0:
                for j in xrange(2, n_classes+1):
                    if variance_combinations[l][j] >= (variance + variance_combinations[i4][j - 1]):
                        lower_class_limits[l][j] = lower_class_limit
                        variance_combinations[l][j] = variance + variance_combinations[i4][j - 1]
 
        lower_class_limits[l][1] = 1.
        variance_combinations[l][1] = variance
 
    return lower_class_limits, variance_combinations

def get_jenks_breaks(data, lower_class_limits, n_classes):
    k = len(data) - 1
    kclass = [0.] * (n_classes+1)
    countNum = n_classes
 
    kclass[n_classes] = data[len(data) - 1]
    kclass[0] = data[0]
 
    while countNum > 1:
        elt = int(lower_class_limits[k][countNum] - 2)
        kclass[countNum - 1] = data[elt]
        k = int(lower_class_limits[k][countNum] - 1)
        countNum -= 1
 
    return kclass
 
def jenks(data, n_classes):
    if n_classes > len(data):
        return
    data.sort()
    lower_class_limits, _ = jenks_matrices(data, n_classes)
    return get_jenks_breaks(data, lower_class_limits, n_classes)
```

One important thing to note about the algorithm is that it requires as input the number of classes to classify into. We will see later that some of the more general clustering algorithms require this as well, but not all of them.

For now, let's see how Jenks does when given the correct number of classes.

```python
breaks = jenks(data, n_clusters)

# Plot results
plt.figure(figsize=(5,2))
plt.scatter(data, labels, c=labels)
plt.yticks(())
for b in breaks:
    plt.axvline(b, color='k');
```

# sklearn.cluster

`scikit-learn` has a number of clustering algorithms, which are described in detail in [the documentation](http://scikit-learn.org/stable/modules/clustering.html). Let's try them all!

## K-Means

The K-Means algorithms aims to find centroids that minimize the distance from points in a cluster and that centroid. It also requires the number of clusters as input.

There is also a more efficient but less accurate method called Mini Batch K-Means, which we also test in this section.

```python
# Do K-Means clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=n_clusters)
kmeans_labels = kmeans.fit_predict(data)

# Do Mini Batch K-Means clustering
from sklearn.cluster import MiniBatchKMeans
minikmeans = MiniBatchKMeans(n_clusters=n_clusters)
minikmeans_labels = minikmeans.fit_predict(data)

# Plot the results
plt.figure(figsize=(5, 6))
plt.subplot(311)
plt.title("Actual clusters")
plt.scatter(data, labels, c=labels)
plt.yticks(())
plt.subplot(312)
plt.title("K-Means")
plt.scatter(data, kmeans_labels, c=kmeans_labels)
plt.yticks(())
plt.subplot(313)
plt.title("Mini Batch K-Means")
plt.scatter(data, minikmeans_labels, c=minikmeans_labels)
plt.yticks(())
plt.tight_layout();
```
