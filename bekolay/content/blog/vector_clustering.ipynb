# One-dimensional clustering

Clustering in many dimensions is a difficult problem with many proposed solutions. In Python, many solutions have been implemented in [scikit-learn](http://scikit-learn.org/stable/modules/clustering.html). How well do these solutions work in one-dimension? Is clustering still valid for one-dimensional data?

```python
# Set up some general imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Test data

To test, we need a vector. We'll use `scikit-learn`'s dataset generating abilities to make a vector with samples from 3 Gaussians.

```python
from sklearn.datasets import make_blobs
n_clusters = 4
n_samples = 40
data, labels = make_blobs(n_features=1, n_samples=n_samples, centers=n_clusters)

plt.figure(figsize=(10,2))
plt.subplot(121)
plt.scatter(data, np.zeros_like(data))
plt.yticks(())
plt.subplot(122)
plt.scatter(data, labels, c=labels)
plt.yticks(());
```

Since we know what Gaussian produced each point, we can separate them in the right subplot, and color them approprately; in actuality, though, what the algorithm will receive is the unlabeled data in the left subplot.

You might want to run the above cell a few times to get data that is reasonably separate, but still with some overlap.

# Kernel Density Estimation

An alternative to clustering for one-dimensional data is to visualize the data as a smooth function, and look at the shape of the function to pick out clusters. While this isn't exactly ground truth, it can give us a good manual alternative to the clustering techniques in order to evaluate them.

One method of approximating a function given samples is [Kernel density esimation](http://en.wikipedia.org/wiki/Kernel_density_estimation). SciPy has an [implementation of Gaussian KDE](http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gaussian_kde.html) in it stats package. Let's see how it does on the data above.

```python
from scipy.stats import gaussian_kde

# Build the Kernel Density Estimator with automatic methods
kde_scott = gaussian_kde(data[:,0], bw_method='scott')
kde_silverman = gaussian_kde(data[:,0], bw_method='silverman')

# Manually play around with the bandwidth
bandwidth = 0.1
kde_manual = gaussian_kde(data[:,0], bw_method=bandwidth)

# What points are we going to estimate?
points = np.linspace(data.min() - 5, data.max() + 5, 100)

# Plot the KDE result
plt.scatter(data, np.zeros_like(data), c=labels)
plt.plot(points, kde_scott(points), lw=2, label="Scott's")
plt.plot(points, kde_silverman(points), lw=2, label="Silverman's")
plt.plot(points, kde_manual(points), lw=2, label="Bandwidth = " + str(bandwidth))
plt.legend(frameon=False, loc='lower center', ncol=3, prop={'size': 10});
```

It seems as though manually specifying the bandwidth is necessary with this kind of data.

# sklearn.cluster

`scikit-learn` has a number of clustering algorithms, which are described in detail in [the documentation](http://scikit-learn.org/stable/modules/clustering.html). Let's try them all!

## K-Means

The K-Means algorithms aims to find centroids that minimize the distance from points in a cluster and that centroid. It also requires the number of clusters as input.

There is also a more efficient but less accurate method called Mini Batch K-Means, which we also test in this section.

```python
# Do K-Means clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=n_clusters)
kmeans_labels = kmeans.fit_predict(data)

# Do Mini Batch K-Means clustering
from sklearn.cluster import MiniBatchKMeans
minikmeans = MiniBatchKMeans(n_clusters=n_clusters)
minikmeans_labels = minikmeans.fit_predict(data)

# Plot the results
plt.figure(figsize=(5, 6))
plt.subplot(311)
plt.title("Actual clusters")
plt.scatter(data, labels, c=labels)
plt.yticks(())
plt.subplot(312)
plt.title("K-Means")
plt.scatter(data, kmeans_labels, c=kmeans_labels)
plt.yticks(())
plt.subplot(313)
plt.title("Mini Batch K-Means")
plt.scatter(data, minikmeans_labels, c=minikmeans_labels)
plt.yticks(())
plt.tight_layout();
```

## Affinity Propagation

Affinity propagation sends messages between pairs of samples until convergence to find clusters in a data set. The n_clusters does not have to be specified as it finds the number of clusters based on the samples given.

A downside of affinity propagation is its complexity - it may take a long time with a large dataset!

```python
# Do Affinity Propagation clustering
from sklearn.cluster import AffinityPropagation
affinity = AffinityPropagation()  # Doesn't require n_clusters!
affinity_labels = affinity.fit_predict(data)

# Plot the results
plt.figure(figsize=(5, 4))
plt.subplot(211)
plt.title("Actual clusters")
plt.scatter(data, labels, c=labels)
plt.yticks(())
plt.subplot(212)
plt.title("Affinity Propagation")
plt.scatter(data, affinity_labels, c=affinity_labels)
plt.yticks(())
plt.tight_layout();
```

## Mean Shift

MeanShift clusters data by estimating "blobs" in a smooth density of points matrix. What is a smooth density of points matrix? Something to do with getting rid of the noise that may get in the way of finding these clusters of blobs.

On the bright side, it doesn't require n_clusters as an input! But it may only work for certain types of data clusters.

```python
# Do Mean Shift clustering
from sklearn.cluster import MeanShift
meanshift = MeanShift()  # Doesn't require n_clusters!
meanshift_labels = meanshift.fit_predict(data)

# Plot the results
plt.figure(figsize=(5, 4))
plt.subplot(211)
plt.title("Actual clusters")
plt.scatter(data, labels, c=labels)
plt.yticks(())
plt.subplot(212)
plt.title("Mean Shift")
plt.scatter(data, meanshift_labels, c=meanshift_labels)
plt.yticks(())
plt.tight_layout();
```
